```python
# Import Dependicies

import pandas as pd
import numpy as np
```

```python
# Read CSVs

csv_202010 = "../Resources/JC-202010-citibike-tripdata.csv"
csv_202011 = "../Resources/JC-202011-citibike-tripdata.csv"
csv_202012 = "../Resources/JC-202012-citibike-tripdata.csv"

bike_data_202010 = pd.read_csv(csv_202010)
bike_data_202011 = pd.read_csv(csv_202011)
bike_data_202012 = pd.read_csv(csv_202012)
```

```python
# Changing gender column to string values

bike_data_202010['gender'] = bike_data_202010['gender'].astype(str)
bike_data_202011['gender'] = bike_data_202011['gender'].astype(str)
bike_data_202012['gender'] = bike_data_202012['gender'].astype(str)
```

```python
# Changing gender column to reflect Female/Male for 202010

gender_202010 = bike_data_202010['gender']
female_202010 = gender_202010.replace('2', 'Female')
male_202010 = female_202010.replace('1', 'Male')
unknown_202010 = male_202010.replace('0', 'Unknown')
bike_data_202010['gender'] = unknown_202010
bike_data_202010.head()
```

```python
# Changing gender column to reflect Female/Male for 202011

gender_202011 = bike_data_202011['gender']
female_202011 = gender_202011.replace('2', 'Female')
male_202011 = female_202011.replace('1', 'Male')
unknown_202011 = male_202011.replace('0', 'Unknown')
bike_data_202011['gender'] = unknown_202011
bike_data_202011.head()
```

```python
# Changing gender column to reflect Female/Male for 202012

gender_202012 = bike_data_202012['gender']
female_202012 = gender_202012.replace('2', 'Female')
male_202012 = female_202012.replace('1', 'Male')
unknown_202012 = male_202012.replace('0', 'Unknown')
bike_data_202012['gender'] = unknown_202012
bike_data_202012.head()
```

```python
# Calculating the age of each biker

current_year = 2021

birth_year_202010 = bike_data_202010['birth year']
age = current_year - birth_year_202010
bike_data_202010['age'] = age

birth_year_202011 = bike_data_202011['birth year']
age = current_year - birth_year_202011
bike_data_202011['age'] = age

birth_year_202012 = bike_data_202012['birth year']
age = current_year - birth_year_202012
bike_data_202012['age'] = age
```
