```python
# Import Dependicies

import pandas as pd
import numpy as np
```

```python
# Read CSVs

csv_2019 = "../Resources/JC-201912-citibike-tripdata.csv"
csv_2020 = "../Resources/JC-202012-citibike-tripdata.csv"

bike_data_2019 = pd.read_csv(csv_2019)
bike_data_2020 = pd.read_csv(csv_2020)
```

```python
# View bike 2019 data

bike_data_2019.head()
```

```python
# View bike 2020 data

bike_data_2020.head()
```

```python
# Checking value types for 2019

bike_data_2019.dtypes
```

```python
# Checking value types for 2020

bike_data_2020.dtypes
```

```python
# Changing gender column to string values

bike_data_2019['gender'] = bike_data_2019['gender'].astype(str)
bike_data_2020['gender'] = bike_data_2020['gender'].astype(str)
```

```python
# Changing gender column to reflect Female/Male for 2019

gender_2019 = bike_data_2019['gender']
female_2019 = gender_2019.replace('2', 'Female')
male_2019 = female_2019.replace('1', 'Male')
unknown_2019 = male_2019.replace('0', 'Unknown')
bike_data_2019['gender'] = unknown_2019
bike_data_2019.head()
```

```python
# Changing gender column to reflect Female/Male for 2020

gender_2020 = bike_data_2020['gender']
female_2020 = gender_2020.replace('2', 'Female')
male_2020 = female_2020.replace('1', 'Male')
unknown_2020 = male_2020.replace('0', 'Unknown')
bike_data_2020['gender'] = unknown_2020
bike_data_2020.head()
```

```python
# Calculating the age of each biker

current_year = 2021

birth_year_2019 = bike_data_2019['birth year']
age = current_year - birth_year_2019
bike_data_2019['age'] = age

birth_year_2020 = bike_data_2020['birth year']
age = current_year = birth_year_2020
bike_data_2020['age'] = age
```

```python
# Rearrange 2019 columns

bike_data_2019_rearranged = bike_data_2019[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_data_2019_rearranged.head()
```

```python
# Rearrange 2020 columns

bike_data_2020_rearranged = bike_data_2019[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_data_2020_rearranged.head()
```

```python
# Rename 2019 columns

bike_data_2019_renamed = bike_data_2019_rearranged.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})

bike_data_2019_renamed.head()
```

```python
# Rename 2020 columns

bike_data_2020_renamed = bike_data_2020_rearranged.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})

bike_data_2020_renamed.head()
```

```python
# Save each cleaned dataset

bike_data_2019_renamed.to_csv("../Outputs/bike_data_2019", index=False, header=True)
bike_data_2020_renamed.to_csv("../Outputs/bike_data_2020", index=False, header=True)
```
